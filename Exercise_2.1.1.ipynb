```python
#Create different scatter plots and graphs for the data in the csv file
sns.pairplot(df)
plt.show()
```

```python
#Choose a number of rows with which you want to work with.
df_subset = df[['area', 'bedrooms', 'price', 'rooms']]

sns.pairplot(df_subset)
plt.show()
```
