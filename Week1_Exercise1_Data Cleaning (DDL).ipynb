```python
#Download all modules needed for the exercise
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Get data and clean it us so they reseamble eachother
df1 = pd.read_csv("Steps_Data.csv", encoding="utf-8", delimiter=";")
df2 = pd.read_csv("Survey_Data.csv", encoding="utf-8")
```

```python
#Merge the two data sets
df_main = pd.merge(df1,df2, on="id")
df_main
```

```python
#Visualising the waist responses in a diagram
visual1 = df_main.plot(y="waist")
```

```python
#Show the highest values in a graph
highest_waist = df_main.sort_values(by="waist",ascending=False)
highest_waist.head(10)
```

```python
#Show lowest valuse in graph based on waist category
lowest_waist = df_main.sort_values(by = 'waist', ascending = True)
lowest_waist.head(10)
```

```python
# removing extreme values
# removing extreme values
def remove_extremes(x):
    if(x > 130.0): 
        return float('NaN')
    if(x < 40.0): 
        return float('NaN')
    else: 
        return x

df_main['waist_corrected'] = df_main['waist'].apply(remove_extremes)
```

```python
waist_corrected = df_main.plot(y="waist_corrected")
```
